# VESPA/EPN-Core services: inspecting some content

In a series of folder/json-files in the current directory samples of EPN-TAP services were fetch.
Data fetching was done with `download_data.py` script.

The goal here is to inspect the content of each service and extract groups and common content; it will be used next to improve Vespa-App' content.

```python
# Load the list of services (currently) available in VESPA's registry
# (kindly provided by Pierre L. Sidaner: file './virtualRegistry.json')
#
import json
with open('virtualRegistry.json') as fp:
    registry_list = json.load(fp)

import pandas
services = pandas.read_json(json.dumps(registry_list['services']), orient='record')
services
```

```python
import sys
import os
import sty
for s in services['schema']:
    if not os.path.isdir(s):
        print('\n'+sty.bg.red + "Service '{!s}' not found".format(s) + sty.bg.rs + '\n')
        continue
    print(sty.fg.blue + '{!s}/'.format(s) + sty.fg.rs)
    for f in os.listdir(s):
        print('\t{!s}'.format(f))
```

Let's snippet a bit of their content

```python
df_master = None
for s in services['schema']:
    if not os.path.isdir(s):
        continue
    f = [f for f in os.listdir(s) if '_0' not in f][0]
    f = os.path.join(s,f)
    with open(f,'r') as fp:
        df = pandas.read_json(fp, orient='record')
        df['schema'] = s
        df = df[:1]
        if df_master is not None:
            df_master = pandas.concat([df_master,df], sort=False)
        else:
            df_master = df
```

```python
len(df_master)
```

```python
cols = df_master.columns.values.tolist()
print(cols)
cols.insert(0, cols.pop(cols.index('schema_epn_core')))
print(cols)
```

```python
df_bla = df_master[cols]
```

```python
df_bla.set_index('schema', drop=True, inplace=True)
```
