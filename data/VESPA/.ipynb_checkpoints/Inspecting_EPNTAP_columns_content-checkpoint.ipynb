# VESPA/EPN-Core services: inspecting some content

In a series of folder/json-files in the current directory samples of EPN-TAP services were fetch.
Data fetching was done with `download_data.py` script.

The goal here is to inspect the content of each service and extract groups and common content; it will be used next to improve Vespa-App' content.

```python
# Load the list of services (currently) available in VESPA's registry
# (kindly provided by Pierre L. Sidaner: file './virtualRegistry.json')
#
import json
with open('virtualRegistry.json') as fp:
    registry_list = json.load(fp)

import pandas
services = pandas.read_json(json.dumps(registry_list['services']), orient='record')
services
```

```python
import sys
import os
import sty
for s in services['schema']:
    if not os.path.isdir(s):
        print('\n'+sty.bg.red + "Service '{!s}' not found".format(s) + sty.bg.rs + '\n')
        continue
    print(sty.fg.blue + '{!s}/'.format(s) + sty.fg.rs)
    for f in os.listdir(s):
        print('\t{!s}'.format(f))
```

Let's snippet a bit of their content

```python
from glob import glob

df_master = None
for s in services['schema']:
    # for each epn_core service/schema in the registry file..
    if not os.path.isdir(s):
        continue
    f = [f for f in glob(os.path.join(s,"*.json")) if '_0' not in f][0]
    # Let's read the retrieved data in the json file..
    print("Reading file: {}".format(f))
    with open(f,'r') as fp:
        df = pandas.read_json(fp, orient='record')
        df['schema'] = s
        df = df[:1]    # <------- Notice that we are taking just the first row of the table!
        if df_master is not None:
            df_master = pandas.concat([df_master,df], sort=False)
        else:
            df_master = df
```

**Looking the table, though, we notice that still many cells are blank, which means that `pandas.DataFrame.isnull()` does not consider _empty strings_ --apparently every field here is 'string'-- as _null values_.** Although it is kind of disappointing, we may actually benefit from that "error"; For instance, the `s_region` field is an important field that gladly we have it here.

So let's just move on with our exploration. Now we will consider, for instance, empty cells as meaningful values to our grouping.

To help the selection of useful columns we will eventually arrive, let's do now a visual and conceptual summary of our table:
* **schema_epn_core**: name of the service/schema, by definition of first importance;
* creation_date: nehhh, we don't really care about that;
* **target_name**: name of the body (a planet, moon, asteroid name) the granule corresponds to;
* **target_class**: class of the body (planet, satellite, asteroid, star) associated;
* spatial_frame_type: basically the system where coordinates/time hold to;
* service_title: not really expressive, will not care about it now, but may be in the future;
* **s_region**: if the corresponding granule has some footprint associated it goes here, we want it;
* processing_level: not interesting;
* **obs_id**: observation ID is always good to have it;
* modification_date: nope;
* **measurement_type**: this is informative, although a bit messy we may find some use;
* **instrument_name**: this is an important information;
* **instrument_host_name**: this is also very important information;
* **granule_uid**: mandatory;
* **granule_gid**: mandatory;
* **dataproduct_type**: this is the type of the datum associated, definetely important;
* release_date: again, I'm not interested in dates for now.

```python
selected_fields = '''schema_epn_core
                     target_name
                     target_class
                     s_region
                     obs_id
                     measurement_type
                     instrument_name
                     instrument_host_name
                     granule_uid
                     granule_gid
                     dataproduct_type'''.split()
```

```python
def agg_df_keywords(df):

    def sRegion_nonempty(series):
        non_empty_vals = sum(series.apply(lambda s:bool(len(s))))
#         empty_vals = sum(not series.str.strip().str.match(''))
#         res = None if empty_vals == len(series) else empty_vals
#         return str(res)
        return non_empty_vals
    
    def fields_unique(series):
        return ';'.join([s.strip() for s in set(series.astype(str).tolist()) if s.strip()])

    df_summ = pandas.Series()
    df_summ['target_name'] = fields_unique(df['target_name'])
    df_summ['target_class'] = fields_unique(df['target_class'])
    df_summ['measurement_type'] = fields_unique(df['measurement_type'])
    df_summ['instrument_name'] = fields_unique(df['instrument_name'])
    df_summ['instrument_host_name'] = fields_unique(df['instrument_host_name'])
    df_summ['dataproduct_type'] = fields_unique(df['dataproduct_type'])
    df_summ['s_region'] = sRegion_nonempty(df['s_region'])
#     df_summ = df.agg({'target_name':fields_unique,
#                         'target_class':fields_unique,
#                         'measurement_type':fields_unique,
#                         'instrument_name':fields_unique,
#                         'instrument_host_name':fields_unique,
#                         'dataproduct_type':fields_unique})#,
#                         's_region':sRegion_nonempty})
    schema = df['schema_epn_core'].unique()
    assert len(schema) == 1
    df_summ.name = schema[0]
    return df_summ
```

```python
# Now we want to check what each service is providing on those fields.
# Fields supposedly providing a limited number of values we want to create the "set" of them:
# - target_name
# - target_class
# - measurement_type
# - instrument_name
# - instrument_host_name
# - dataproduct_type
# Fields which values can vary more or less freely we want to check if they can be empty:
# - s_region
# - obs_id
# - granule_uid
# - granule_gid

from IPython.display import display,HTML

df_agg = None

s_list = services['schema']#[0:3]
for s in s_list:

    if not os.path.isdir(s):
        print('\n' + sty.bg.red + '{!s}'.format(s) + sty.bg.rs + '\n')
        continue
        
    f = [f for f in glob(os.path.join(s,"*.json")) if '_0' not in f][0]
    print(sty.fg.blue + '{!s}'.format(s) + sty.fg.rs)
    
    with open(f,'r') as fp:
        df = pandas.read_json(fp, orient='record')
    df = df[selected_fields]
    
    display(HTML(df.head(10).to_html()))

    s_agg = agg_df_keywords(df)
    print(s_agg.to_frame())
    
    if df_agg is not None:
        df_agg = pandas.concat([df_agg,s_agg.to_frame()], axis=1, sort=False)
    else:
        df_agg = s_agg.to_frame()

    print('\n')
```

```python
df_agg_T = df_agg.T
df_agg_T
```

### Actions

Let's work with the following top-level groups:
* From `target_class`
* From `target_name`

```python
def splitDataFrameList(df,target_column,separator):
    ''' df = dataframe to split,
    target_column = the column containing the values to split
    separator = the symbol used to perform the split
    returns: a dataframe with each entry for the target column separated, with each element moved into a new row. 
    The values in the other columns are duplicated across the newly divided rows.
    '''
    def splitListToRows(row,row_accumulator,target_column,separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)
    new_rows = []
    df.apply(splitListToRows,axis=1,args = (new_rows,target_column,separator))
    new_df = pandas.DataFrame(new_rows)
    return new_df[df.columns]
```

```python
index_name = df_agg_T.index.name
df_agg_T.reset_index(inplace=True)

df_agg_T_CLASS = splitDataFrameList(df_agg_T, 'target_class', ';')
df_agg_T_CLASS.set_index(index_name or 'index', inplace=True)
if not index_name:
    df_agg_T_CLASS.index.name = None
df_agg_T_CLASS
```

```python
index_name = df_agg_T_CLASS.index.name
df_agg_T_CLASS.reset_index(inplace=True)

df_agg_T_CLASS_NAME = splitDataFrameList(df_agg_T_CLASS, 'target_name', ';')
df_agg_T_CLASS_NAME.set_index(index_name or 'index', inplace=True)
if not index_name:
    df_agg_T_CLASS_NAME.index.name = None
df_agg_T_CLASS_NAME
```
